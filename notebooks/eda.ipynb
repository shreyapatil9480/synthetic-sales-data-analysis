Which customers are likely to churn?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/customer_tenure.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
target = 'churned'
df[target].value_counts(normalize=True).plot(kind='bar', title='Target rate')
plt.tight_layout()
```

## Segment breakdown

```python
numeric = df.select_dtypes('number').columns
df[numeric].hist(figsize=(8,6))
plt.tight_layout()
```
